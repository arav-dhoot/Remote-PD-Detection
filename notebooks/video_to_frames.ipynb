```python
os.makedirs(destination_path, exist_ok=True)
for coded_folder in os.listdir(video_path):
    if not os.path.isdir(os.path.join(video_path, coded_folder)): os.remove(os.path.join(video_path, coded_folder))
    os.makedirs(os.path.join(destination_path, coded_folder), exist_ok=True)
    for severity in os.listdir(os.path.join(video_path, coded_folder)):
        if not os.path.isdir(os.path.join(video_path, coded_folder, severity)): os.remove(os.path.join(video_path, coded_folder, severity))
        os.makedirs(os.path.join(destination_path, coded_folder, severity), exist_ok=True)
        for video in os.listdir(os.path.join(video_path, coded_folder, severity)):
            os.makedirs(os.path.join(destination_path, coded_folder, severity, video), exist_ok=True)
            video_capture = cv2.VideoCapture(os.path.join(video_path, coded_folder, severity, video))
            frame_count = 0
            while True:
                ret, frame = video_capture.read()
                if not ret: break
                if frame_count % 10 == 0:
                    output_file = os.path.join(destination_path, coded_folder, severity, video, f'{frame_count}.jpg')
                    cv2.imwrite(output_file, frame)
                frame_count += 1
            video_capture.release()
            print(f"Frames extracted: {frame_count}")
            print(f"Extraction complete. Files saved at {os.path.join(destination_path, coded_folder, severity, video)}")
```
