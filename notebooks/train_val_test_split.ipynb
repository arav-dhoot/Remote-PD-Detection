```python
import os
import random
import shutil
```

```python
base_folder = 
sub_folders = ['MILD', 'MODERATE', 'NORMAL', 'SEVERE']
subsets = ['Train', 'Validation', 'Test']

split_percentages = [0.7, 0.1, 0.2]
```

```python
for subset in subsets:
    subset_folder = os.path.join(base_folder, subset)
    os.makedirs(subset_folder, exist_ok=True)
    for sub_folder in sub_folders:
        os.makedirs(os.path.join(subset_folder, sub_folder), exist_ok=True)
```

```python
image_files = []
for sub_folder in sub_folders:
    sub_folder_path = os.path.join(base_folder, sub_folder)
    for root, dirs, files in os.walk(sub_folder_path):
        for file in files:
            if file.lower().endswith(('.png', '.jpg', '.jpeg', '.gif', '.bmp')):
                image_files.append(os.path.join(root, file))
            
random.shuffle(image_files)
```

```python
split_sizes = [int(len(image_files) * p) for p in split_percentages]
train_set = image_files[:split_sizes[0]]
validation_set = image_files[split_sizes[0]:split_sizes[0] + split_sizes[1]]
test_set = image_files[split_sizes[0] + split_sizes[1]:]

for subset, image_set in zip(subsets, [train_set, validation_set, test_set]):
    for image_path in image_set:
        sub_folder = os.path.basename(os.path.dirname(image_path))
        destination_folder = os.path.join(base_folder, subset, sub_folder)
        shutil.copy(image_path, destination_folder)
```
