```python
import os
```

```python
def clean_folders(location):
    for root, dirs, files in os.walk(location):
        for file in files:
            if file == ".DS_Store":
                file_path = os.path.join(root, file)
                os.remove(file_path)
                print(f"Deleted: {file_path}")
```

```python
clean_folders()
```
